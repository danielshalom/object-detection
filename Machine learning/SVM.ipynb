# SVM

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import SVC
```

# Linear SVM

```python
x1,x2,y=np.loadtxt('SVM_data1.csv',delimiter=',',unpack=True)

plt.plot(x1[y==-1],x2[y==-1],'ob')
plt.plot(x1[y==1],x2[y==1],'or')
plt.show()
```

```python
itr=2
n =0
C=10
while n < itr:
 x1,x2,y=np.loadtxt('SVM_data1.csv',delimiter=',',unpack=True)
 clf=SVC(C=C,kernel='linear')
 X=np.array((x1,x2))
 X=X.transpose()
 clf.fit(X,y)
 xx, yy= np.meshgrid(np.linspace(-1,1,500), np.linspace(-1,1,500))

 Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
 Z = Z.reshape(xx.shape)

 plt.contourf(xx,yy,Z,100, alpha=0.3, cmap=plt.cm.bwr, antialiased=True)
 plt.contour(xx,yy,Z, [0])
 plt.contour(xx,yy,Z, [1])
 plt.contour(xx,yy,Z, [-1])
 plt.plot(x1[y==-1],x2[y==-1],'ob')
 plt.plot(x1[y==1],x2[y==1],'or')

 cnt=clf.predict(X)
 errors = cnt-y
 plt.plot(x1[errors != 0], x2[errors != 0], 'og')
 print('Errors : ' ,len(errors[errors != 0]))
 plt.show()
 C=1000
 n+=1
```

# SVM with polynomial kernel

```python
x1,x2,y=np.loadtxt('SVM_data2.csv',delimiter=',',unpack=True)

plt.plot(x1[y==-1],x2[y==-1],'ob')
plt.plot(x1[y==1],x2[y==1],'or')
plt.show()
```

```python
x1,x2,y=np.loadtxt('SVM_data2.csv',delimiter=',',unpack=True)

clf=SVC(C=40,kernel='poly', degree=2)
X=np.array((x1,x2))
X=X.transpose()
clf.fit(X,y)
xx, yy= np.meshgrid(np.linspace(-1,1,500), np.linspace(-1,1,500))

Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx,yy,Z,100, alpha=0.3, cmap=plt.cm.bwr, antialiased=True)
plt.contour(xx,yy,Z, [0])
plt.contour(xx,yy,Z, [1])
plt.contour(xx,yy,Z, [-1])
plt.plot(x1[y==-1],x2[y==-1],'ob')
plt.plot(x1[y==1],x2[y==1],'or')

cnt=clf.predict(X)
errors = cnt-y
plt.plot(x1[errors != 0], x2[errors != 0], 'og')
print('Errors : ' ,len(errors[errors != 0]))
plt.show()
```

# SVM with RBF kernel

```python
x1,x2,y=np.loadtxt('SVM_data3.csv',delimiter=',',unpack=True)

plt.plot(x1[y==-1],x2[y==-1],'ob')
plt.plot(x1[y==1],x2[y==1],'or')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.svm import SVC  
x1,x2,y=np.loadtxt('SVM_data3.csv',delimiter=',',unpack=True)

X=np.array((x1,x2))
X=X.transpose()
xx, yy= np.meshgrid(np.linspace(0,1,500), np.linspace(0,1,500))

clf=SVC(C=400000,kernel='rbf')
clf.fit(X,y)

Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx,yy,Z,100, alpha=0.7, cmap=plt.cm.bwr, antialiased=True)
plt.contour(xx,yy,Z, [0])
plt.contour(xx,yy,Z, [1])
plt.contour(xx,yy,Z, [-1])

plt.plot(x1[y==-1],x2[y==-1],'ob')
plt.plot(x1[y==1],x2[y==1],'or')

cnt=clf.predict(X)

errors = cnt-y
plt.plot(x1[errors != 0], x2[errors != 0], 'og')
print('Errors : ' ,len(errors[errors != 0]))
plt.show()
```
